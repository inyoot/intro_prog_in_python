## Branching examples

```python
# Ask an integer as input
x = int(input("Please enter an integer: "))

# Calculate the modolus two of x
m = x % 2

# Determine if the number is even or odd
if m == 0:
    print('Even number')
else:
    print('Odd number')
```

```python
x = int(input("Please enter an integer: "))
m = x % 3
if m == 0:
    print('Remainder = 0')
elif m == 1:
    print('Remainder = 1')
else:
    print('Remainder = 2')
```

## Guessing Game

```python
#%%timeit
upper_limit = 2**4
secret_number = upper_limit-3

for num in range(upper_limit):
    if num == secret_number:
        print(num)
        break
```

```python
%%timeit
upper_limit = 2**20
secret_number = upper_limit - 1

for num in range(upper_limit):
    if num == secret_number:
        print(num)
        break
```

```python
%%timeit
upper_limit = 2**4
secret_number = 1
left_limit = 0
right_limit = upper_limit
guess = 0

while (left_limit+1) < right_limit:
    mid_point = left_limit + (right_limit - left_limit)//2
    #print(mid_point)
    if secret_number < mid_point:
        right_limit = mid_point
        guess = (guess << 1) + 0
    else:
        left_limit = mid_point
        guess = (guess << 1) + 1
    #print(left_limit,right_limit,guess)
print(guess)
```
